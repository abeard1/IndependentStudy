### Team Schedule Network Representation

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# without eurocup/euroleague
from sportsref import euro

year = 2015
LEAGUE_IDS = ['greek-basket-league', 'spain-liga-acb', 'italy-basket-serie-a','france-lnb-pro-a']

schedules = []
for league in LEAGUE_IDS:
        season = euro.Season(year, league)
        schedules.append(season.schedule(kind='R'))
    
schedule_df = pd.concat(schedules)
schedule_df.reset_index(drop=True)

schedule_df.head()
```

```python
G = nx.from_pandas_dataframe(schedule_df, 'visitor_team_name', 'home_team_name')

#pos = nx.spring_layout(G)
# will try to make positions consistent

plt.figure(figsize=(4,4))

nx.draw(G, alpha=0.5, node_size=50, node_color='r', with_labels=False)

plt.title('2014-2015 Schedule Network Representation - without EuroCup/EuroLeague')
plt.show()
```
