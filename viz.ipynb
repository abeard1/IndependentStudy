### Team Schedule Network Representation

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_df = pd.read_csv('raw_data/raw_data_2015.csv')
G = nx.from_pandas_dataframe(raw_df, 'team_id', 'opp_id')

plt.figure(figsize=(4,4))

nx.draw(G, alpha=0.5, node_size=50, node_color='r', with_labels=False)

plt.title('2014-2015 Schedule Network Representation')
plt.show()
```
